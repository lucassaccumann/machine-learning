# Clustering Aglomerativo

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import linkage
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_matrix = [[0, 2,6,10,9],
               [2, 0,5, 9,8],
               [6, 5,0, 4,5],
               [10,9,4, 0,3],
               [9, 8,5, 3,0]]
print(data_matrix)
model = AgglomerativeClustering(affinity='precomputed', n_clusters=2, linkage='complete').fit(data_matrix)
print(model.labels_)
```

```python
linkage_matrix = linkage(data_matrix, "single")
dendrogram(linkage_matrix, color_threshold=1, labels=["1","2","3","4","5"], show_leaf_counts=True)
plt.title=("Dendrograma")
plt.show()
```

```python
dataset = pd.read_csv('agrupamento_Q1.csv')
```

```python
dataset.iloc[:,:].head()
```

```python
plt.figure(figsize=(10, 7))  
dend = shc.dendrogram(shc.linkage(dataset, method='ward'))  
```

```python
model2 = AgglomerativeClustering(n_clusters=5, linkage='complete').fit(dataset)
print(model2.labels_)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(dataset.iloc[:,0], dataset.iloc[:,1], c=model2.labels_, cmap='rainbow')  
plt.figure(figsize=(10, 7))  
plt.scatter(dataset.iloc[:,0], dataset.iloc[:,2], c=model2.labels_, cmap='rainbow')  
plt.figure(figsize=(10, 7))  
plt.scatter(dataset.iloc[:,0], dataset.iloc[:,3], c=model2.labels_, cmap='rainbow')  
plt.figure(figsize=(10, 7))  
plt.scatter(dataset.iloc[:,1], dataset.iloc[:,2], c=model2.labels_, cmap='rainbow')
plt.figure(figsize=(10, 7))  
plt.scatter(dataset.iloc[:,1], dataset.iloc[:,3], c=model2.labels_, cmap='rainbow')
plt.figure(figsize=(10, 7))  
plt.scatter(dataset.iloc[:,2], dataset.iloc[:,3], c=model2.labels_, cmap='rainbow')
```
